```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Win 10\Downloads\traffic project\traffic accident.csv")
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['crash_hour'],axis=1)
y = df['crash_hour']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.select_dtypes(include="number").corr()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include='number').corr(), annot=True, cmap="YlGnBu")
```

```python
train_data['injuries_total'] = np.log(train_data['injuries_total']+1)
train_data['injuries_fatal'] = np.log(train_data['injuries_fatal']+1)
train_data['injuries_no_indication'] = np.log(train_data['injuries_no_indication']+1)


```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data.crash_type.value_counts()
```

```python
train_data = train_data.join(pd.get_dummies(train_data.crash_type)).drop(['crash_type'],axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.select_dtypes(include='number').corr(),annot=True, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x="num_units",y="injuries_total",data=train_data, hue="crash_hour",palette="coolwarm")
```
